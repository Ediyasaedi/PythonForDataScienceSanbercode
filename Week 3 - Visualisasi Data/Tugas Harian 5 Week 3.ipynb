```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../csv/vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
mean_score = df.groupby(['Genre'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].agg('mean')
mean_score
```

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# CARA I
import numpy as np

plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(16, 8))
x = np.arange(0, 12, 3)
x0 = 0

# cara looping iterrows 1
for row in mean_score.iterrows():
    ax.bar(x+x0, row[1], width=0.2, label=row[0])
    x0 += 0.2

# # cara looping iterrows 2
# for rowname, row in mean_score.iterrows():
#     ax.bar(x+x0, row, width=0.2, label=rowname)
#     x0 += 0.2

plt.xticks(x+1, mean_score.columns)

ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')

fig.legend()
plt.show()
```

```python
# CARA II
import numpy as np

plt.style.use('ggplot')

genre = np.sort(df['Genre'].unique())

fig, ax = plt.subplots(figsize=(16, 8))
x = np.arange(0, 24, 6)

# looping berdasarkan Genre
for idxg, g in enumerate(genre):
    ax.bar(x+0.3*idxg, mean_score.loc[g], width=0.3, label=g)

# cara lain memberi label pada sumbu x
ax.set_xticks(x+1)
ax.set_xticklabels(mean_score.columns)

ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')

fig.legend()
plt.show()
```

<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
